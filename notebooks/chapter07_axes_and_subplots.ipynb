## Figure
In Matplotlib, a **Figure** acts as a container that holds all the plot elements. Think of it as the canvas on which your plots will be drawn. Creating a Figure in matplotlib is usually the first step in plotting data. You do this by calling `plt.figure()`, which initializes a new Figure object. This object provides the foundation upon which you'll build your visualization, add axes, labels, and other elements.

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

def display_image(image_path, new_size):
    # Open the image file
    image = Image.open(image_path)

    # Resize the image
    resized_image = image.resize(new_size)

    # Convert the resized image to an array
    img = np.asarray(resized_image)

    # Display the resized image
    imgplot = plt.imshow(img)
    plt.axis('off')  # Hide the axis
    plt.show()

display_image('../images/axes1.PNG', (500, 400))
```

## Axes

An Axes in Matplotlib is a single plot or graph that forms the core building blocks for creating data visualizations. It represents a distinct area of a figure where data is plotted and includes functionalities for generating a variety of plot types, such as line plots, bar charts, and more. 

You can run the following code in jupyter notebook as usual.

```python
# Create a figure
fig = plt.figure(figsize=(6, 4), facecolor="teal")

# Add a single axes to the figure
# The (1, 1, 1) mean (rows, columns, subplot index)
ax = fig.add_subplot(1, 1, 1)

# Optionally, set the axes background color
ax.set_facecolor("deeppink")
plt.savefig("../images/axes03.png")

plt.show()
```

The `Figure` is the entire space with the `teal` background, while the `axes` is the area with a deep pink background where data will be plotted. By default, matplotlib has added some ticks and tick labels which will change when we add some data.

Let's add a plot to this axes. We will add a simple line chart to it.

```python
import warnings

warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
import pandas as pd

# read data
nvidia = pd.read_csv("../data/NVDA.csv", parse_dates=["Date"])

# Create a figure
fig = plt.figure(figsize=(6, 4), facecolor="teal")

# add a line chart
ax = fig.add_subplot(1, 1, 1)
ax.plot(nvidia["Date"], nvidia["Adj Close"])
ax.set_title("Nvidia Stock Prices (2019 - 2024)")

# set the axes background color
ax.set_facecolor("deeppink")
plt.savefig("../images/axes_extra.png")

plt.show()
```

## Axis

The Axis is the number line we usually see on graphs. It controls the graph limits, tick marks, and their labels. An Axes object has at least 2 axis ( the x-axis and the y-axis) for 2D plots and 3 axis for 3D plots.

Let's customize the x and the y-axis of the plot.

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Create a figure
fig = plt.figure(figsize=(6, 4), facecolor="teal")

# add a line chart
ax = fig.add_subplot(1, 1, 1)
ax.plot(nvidia["Date"], nvidia["Adj Close"])
ax.set_title("Nvidia Stock Prices (2019 - 2024)")


# Customize the x-axis to show every 2 years
ax.xaxis.set_major_locator(mdates.YearLocator(2))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
ax.set_xlabel("Year")

# customize the y-axis to show $200 interval
ax.set_yticks(range(0, 701, 200))
ax.set_ylabel("Adj Close Price")


# set the axes background color
ax.set_facecolor("deeppink")

plt.savefig("../images/axes05.png")

# show plot
plt.show()
```

Now, we have a customized x and y-axis.I hope these examples clear any doubt you have about Figures, Axes, and Axis in Matplotlib. Now, let's tackle one another very confusing thing in Matplotlib.

## Subplots

Subplots in Matplotlib are a way to arrange multiple axes (plots) within a single figure. Each subplots represents an individual plot, with its own axes and chart type. Subplots are very useful for comparing data, showing trends, or highlighting relationship between different datasets. 

When you create subplots, you are essentially divide the figure into a grid, where each cell of the grid contains its own axes object with a distinct plot. The grid is defined by rows and columns, allowing for a flexible layout of your visualization. There are several ways to create subplots in matplotlib, but one of the most common method is using the `plt.subplots()` function. 

In essence, an Axes is about the `what` of your data visualization (What you are plotting), while subplots are about the `how`( how you are organizing multiple plots in relation to each other).

Let's see an example of subplots in matplotlib. 


```python
import matplotlib.pyplot as plt

# Create a figure
fig = plt.figure(figsize=(7, 4), facecolor="teal")

# Add the first subplot to the figure. This creates an axes
# in the left half of the figure.
ax1 = fig.add_subplot(1, 2, 1)  # (1 row, 2 columns, 1st subplot)
ax1.set_facecolor("deeppink")
ax1.set_title("First Axes")

# Add the second subplot to the figure, creating another axes
# in the right half of the figure.
ax2 = fig.add_subplot(1, 2, 2)  # (1 row, 2 columns, 2nd subplot)
ax2.set_facecolor("dodgerblue")
ax2.set_title("Second Axes")

plt.savefig("../images/axes04.png")

plt.show()
```
