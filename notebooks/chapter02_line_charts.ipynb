# Line Charts

A line chart is used to show how the value of a variable changes over time. The x-axis of a line chart represents the time frame like months or years. The y-axis represents the value that we are interested in measuring like revenue or number of visitors to a website. We can also create multiple line charts to compare related groups over time like the population of India, China, and US in the past 20 years.

Line charts are particularly useful for identifying trends in data over time. The continuous line makes it easy to see how values increase, decrease, or remain stable during a time frame.


## Read Data 

For data manipulation and analysis we will use the Pandas library. Let's import it and read a dataset using the `read_csv()` function. We will also convert the `Date` column to `datetime` using the `parse_dates` parameter. 

```python
# silence warnings
import warnings

warnings.filterwarnings("ignore")

# import pandas
import pandas as pd

# read the data into pandas DataFrame
nvidia = pd.read_csv("../data/NVDA.csv", parse_dates=["Date"])
nvidia.head()
```



## Create a Line Chart

To create a Line Chart in Matplotlib, we use the [`plt.plot()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html) function. And we use the [`plt.show()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.show.html) function to show the figure. 

Before we create any kind of chart in matplotlib, we need to first import the matplotlib library. Let's create a Line Chart, We will plot the `Date` column on the x-axis and the `Adj Close` column on the y-axis.

```python
# import matplotlib library
import matplotlib.pyplot as plt

# create a line chart
plt.plot(nvidia["Date"], nvidia["Adj Close"])
plt.show()
```

## Add Title and Axis labels

Let's add a title to the plot using the [`plt.title()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html) and x and y axis labels using [`plt.xlabel()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html) and [`plt.ylabel()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html) respectively. 

```python
# line chart with title and axis labels
plt.plot(nvidia["Date"], nvidia["Adj Close"])
plt.title("Nvidia Stock Prices (2019-2024)")
plt.xlabel("Year")
plt.ylabel("Adj Close Price")
plt.show()
```

We can see that the share price of Nvidia were constantly going up from 2019-22 and then it started to go down between 2022-23 and then started to go up again and went very high at the beginning of 2024. The reason for such a hike in Nvidia stock prices is the result of the explosion of Artificial Intelligence and their revenue from selling `GPUs` which are needed to train large machine learning models and Nvidia controls more than 95% of the GPU market

## Multiple Line charts

To create multiple line charts on the same figure, we just need to add another `plt.plot()` function. This will instruct matplotlib to create another line chart that shares the same x and y axis as the first figure. 

Let's read some more data for this.

```python
# read data
netflix = pd.read_csv("../data/NFLX.csv", parse_dates=["Date"])
tesla = pd.read_csv("../data/TSLA.csv", parse_dates=["Date"])
```

We will create a line chart for Nvidia, netflix, and Tesla on the same plot.

```python
# create multiple line charts
plt.plot(nvidia["Date"], nvidia["Adj Close"])

plt.plot(netflix["Date"], netflix["Adj Close"])

plt.plot(tesla["Date"], tesla["Adj Close"])

plt.title("Stock Prices (2019-2024)")
plt.xlabel("Year")
plt.ylabel("Adj Close Price")
plt.show()
```

## Add a Legend

Although we plotted all three lines on the graph, it is not clear which line corresponds to which company. To solve this problem, we need to add a legend to the plot. First, we will add labels to each line chart and then use the [`plt.legend()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html) function to add a legend. 

```python
# multiple line charts with legend
plt.plot(nvidia["Date"], nvidia["Adj Close"], label="Nvidia")

plt.plot(netflix["Date"], netflix["Adj Close"], label="Netflix")

plt.plot(tesla["Date"], tesla["Adj Close"], label="Tesla")

plt.title("Stock Prices (2019-2024)")
plt.xlabel("Year")
plt.ylabel("Adj Close Price")
plt.legend()
plt.show()
```

## Line style and color

Instead of using the default color and line style, we can apply our styles. To change the Line style use the `linestyle` and to change the color use the `color` parameter.

```python
# Line chart with custom line style and color
plt.plot(tesla["Date"], tesla["Adj Close"], label="Tesla", linestyle=":", color="blue")

plt.plot(
    nvidia["Date"],
    nvidia["Adj Close"],
    label="Nvidia",
    linestyle="--",
    color="seagreen",
)

plt.plot(
    netflix["Date"],
    netflix["Adj Close"],
    label="Netflix",
    linestyle="-.",
    color="crimson",
)

plt.title("Stock Prices (2019-2024)")
plt.xlabel("Year")
plt.ylabel("Adj Close Price")
plt.legend()
plt.show()
```

I think the plot with solid lines looked much better than this version. There are many other parameters you can use to change the appearance of a line chart, please check the [`plt.plot()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html) document.

# Style sheets 

There are also various style sheets available to change the appearance of your plots. You can list all the styles available using `plt.style.available` attribute. You can also see the examples of them here - [`stylesheet examples`](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html)

```python
# uncomment and run the below line to see
# print(plt.style.available)
```

To add a style use the `plt.style.use("stylesheet_name")`

```python
# Add a style sheet in matplotlib
plt.style.use("seaborn-v0_8-dark")

plt.plot(tesla["Date"], tesla["Adj Close"], label="Tesla", color="blue")

plt.plot(nvidia["Date"], nvidia["Adj Close"], label="Nvidia", color="seagreen")

plt.plot(netflix["Date"], netflix["Adj Close"], label="Netflix", color="crimson")

plt.title("Stock Prices (2019-2024)")
plt.xlabel("Year")
plt.ylabel("Adj Close Price")
plt.legend()
plt.show()
```

# Figure size 

To change the size of a figure use the `figsize` parameter of [`plt.figure()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html) function. The figsize parameter takes a tuple of (width, height) in inches. 

```python
# Change figure size
plt.figure(figsize=(7, 5))

plt.plot(tesla["Date"], tesla["Adj Close"], label="Tesla", color="blue")

plt.plot(nvidia["Date"], nvidia["Adj Close"], label="Nvidia", color="seagreen")

plt.plot(netflix["Date"], netflix["Adj Close"], label="Netflix", color="crimson")

plt.title("Stock Prices (2019-2024)")
plt.xlabel("Year")
plt.ylabel("Adj Close Price")
plt.legend()
plt.show()
```

# Save a figure 

To save a matplotlib figure use the [`plt.savefig("filepath/image.png")`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html) function before calling `plt.show()`. You can also use other image extensions like jpeg, svg, pdf, etc.

```python
plt.figure(figsize=(7, 5))

plt.plot(tesla["Date"], tesla["Adj Close"], label="Tesla", color="blue")

plt.plot(nvidia["Date"], nvidia["Adj Close"], label="Nvidia", color="seagreen")

plt.plot(netflix["Date"], netflix["Adj Close"], label="Netflix", color="crimson")

plt.title("Stock Prices (2019-2024)")
plt.xlabel("Year")
plt.ylabel("Adj Close Price")
plt.legend()
# save it in the image folder
plt.savefig("../images/stock_prices_2019_2024.png")
plt.show()
```

2. Create a Line chart of Men's and Women's ages at marriage overtime on the same figure.
3. Create a new column, `age_diff`, by subtracting women's age from men's age at marriage.
4. Create a Line chart of age differences over time.


## Summary

- To create a Line Chart in matplotlib use the `plt.plot()` function.
- To add a title to the plot use `plt.title()`.
- For x and y axis label use the `plt.xlabel()` and `plt.ylabel()` respectively.
- Multiple line charts can be added to the same figure by adding another `plt.plot()` function.
- To add a legend, first specify `label` for each plot and then use `plt.legend()`.
- We can change the line style, and color using `linestyle`, and `color` parameters of `plt.plot()`
- To use a style sheet in matplotlib use `plt.style.use("stylesheet_name")`.
- To change the figure size use the `figsize=(width, height)` of `plt.figure()`.
- To save a figure use the `plt.savefig("filepath/image.png")`.

## Solution 

### Exercise 2.1

```python
# ignore warnings
import warnings

warnings.filterwarnings("ignore")

# import libraries
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-dark")

# Read the marriage-age-2016.txt in pandas
marriage = pd.read_csv("../data/marriage-age-2016.txt", sep="\t")

# Create a new column age_diff
marriage["age_diff"] = marriage["Men"] - marriage["Women"]

# Line chart of men's and women's ages at marriage
plt.plot(marriage["Year"], marriage["Men"], label="Men", color="crimson")

plt.plot(marriage["Year"], marriage["Women"], label="Women", color="seagreen")

plt.title("Age at Marriage By Gender (1890 - 2016)")
plt.xlabel("Year")
plt.ylabel("Age at Marriage")
plt.legend()
plt.savefig("../images/line08.png")
plt.show()
```

```python
# Line chart of men's and women age difference.
plt.plot(marriage["Year"], marriage["age_diff"], color="crimson")

plt.title("Age Difference at Marriage (1890 - 2016)")
plt.xlabel("Year")
plt.ylabel("Age difference at Marriage")
plt.savefig("../images/line09.png")
plt.show()
```
